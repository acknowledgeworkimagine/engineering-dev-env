```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math

# Setup
psi_values = [600, 1000, 1500, 2000, 2500]
torque_at_2500psi_nm = 135.6
psi_ref = 2500
rpm_range = np.linspace(600, 1800, 200)

plt.figure(figsize=(10, 6))

# Loop through pressures
for psi in psi_values:
    torque = torque_at_2500psi_nm * psi / psi_ref
    torque_curve = np.full_like(rpm_range, torque)

    # Dynamically formatted LaTeX math display
    latex_eq = (
        rf"\text{{Torque}}_{{{psi}\,\text{{psi}}}} = 135.6 \times \frac{{{psi}}}{{2500}} = {torque:.1f}\,\text{{Nm}}"
    )
    display(Math(latex_eq))

    # Plot torque curve
    plt.plot(rpm_range, torque_curve, label=f'{psi} psi ({torque:.1f} Nm)')

# Finalize plot
plt.title('Motor Torque vs RPM at Different Pressures')
plt.xlabel('Speed (RPM)')
plt.ylabel('Torque (Nm)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
